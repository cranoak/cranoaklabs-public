```python
# Floor Division.  Ever/ything but the remainder
print(f'45/4 = {45/4}')
print(f'45//4 = {45//4}')
print(f'45%4 = {45%4}')
```

```python
# Floor Division.  Ever/ything but the remainder
numerator = 45
denominator = 4
remainder = 45%4
base = 45//4

print(f'{numerator}/{denominator} = {numerator/denominator}') 
print(f'{numerator}//{denominator} = {numerator//denominator}') 
print(f'45%4 = {45%4}')
print(f'base + remainder/denominator: {base + (remainder/denominator)}')
```
