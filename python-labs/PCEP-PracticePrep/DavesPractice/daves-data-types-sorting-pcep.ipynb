```python
# nums = [1,2,3,4]
###### finish for loop so it produces lnames below
# lnames= [(0, 'andy'), (1, 'candy'), (2, 'mandy'), (3, 'sandy')]

names = ['andy', 'candy', 'mandy', 'sandy']
nums = [i for i in range(len(names))]
print(nums)
# build tuple within list
lnames = []

#
for i in nums:
    #print(names[i-1])
    lnames.append((i, names[i-0]))
##### WHAT PRINTS BELOW
print(f'lnames= {lnames}')
print(f'type(lnames= {type(lnames)}')
print(f'type(lnames[0]= {type(lnames[0])}')
    
```

```python
####  Sort lnames based on key so lnames becomes:
#lnames = lnames= [(0, 'andy'), (2, 'candy'), (1, 'mandy')]

names = ['andy', 'mandy', 'candy']
nums = [i for i in range(len(names))]
print(nums)
# build tuple within list
lnames = []
for i in nums:
    lnames.append((i,names[i]))
print("sorted based on alphabet order of names")
print(f'lnames= {lnames}')
print(f'type(lnames) is {type(lnames)} and type(lnames[0]) {type(lnames[0])}')
```
